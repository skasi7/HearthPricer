```python
import numpy
import os.path
import pandas

import hearthpricer.card_loader
import hearthpricer.card_pricer

all_sets_filename = os.path.join('data', 'AllSets.json')

card_list = hearthpricer.card_loader.load(all_sets_filename) #, include_class_cards=False)
len(card_list)
```

```python
minions_df = pandas.DataFrame(card_list)
minions_df.info()
```

```python
coeffs = hearthpricer.card_pricer.pricing(minions_df, columns=[
        u'attack', u'charge', u'divine shield', u'health', u'stealth', u'taunt', u'windfury',
        'intrinsic'], debug=True)
```

```python
card_value = coeffs[-1][0]
minions_df['value'] = (minions_df['price'] - minions_df['cost']) / (minions_df['cost'] + card_value)
minions_df[['name', 'cost', 'price', 'attack', 'health', 'windfury', 'value']].sort(
    'value', ascending=False)
```
